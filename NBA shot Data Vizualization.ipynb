Hello! Welcome to our data visualization project for CS 42. Today we will be investigating the shot strategies of the three top teams in the NBA in the past 2 decades during their prime. The three teams we will be  investigating are the LA Lakers (2000-2003), the Miami Heat (2010-2013), and the Golden State Warriors (2014-2019). We hope to gain some insight on how the strategies of top teams have changed over time!

Of course, we need to get our data somehow. Luckily, the NBA keeps super accurate data about every shot taken in regulation NBA games. They take all this data so each game can be simulated afterwards and teams can review their game to improve their strategy! The NBA's official API is really annoying to access and requires a lot of random arguments. Fortunately for us, there are very smart people who have made a library called *nba_api* that allows us to easily access the NBA's shot data. The specific api call from the library that we want is called *shotchartdetail*.

Below are the imports that we used for this project. When running this notebook you might have to use **pip install (library)** a few times.

The first type of visualization we will do is that of the accuracy of shots. As an example, we will just try using a scatter plot first. We will only try this for the Lakers (you will see why soon).

```python
# Displays scatterplot of shots overlayed over a regulation NBA court.


dfx = lakers_result.loc[(lakers_result.EVENT_TYPE == "Missed Shot")]
dfy = lakers_result.loc[(lakers_result.EVENT_TYPE == "Made Shot")]
plt.scatter(dfx.LOC_X, dfx.LOC_Y, c='coral')
plt.scatter(dfy.LOC_X, dfy.LOC_Y, c='lightblue')
plt.gcf().set_size_inches((15, 20))
court = Image.open("nba_court.jpeg")
plt.imshow(court, extent=[-250,250,-50,420])
plt.show()
```

The red shots represent misses and the blue represents made shots. Clearly, this doesn't really help us much since everything is kind of jumbled and you can't really conceptualize a sea of dots. It does look cool though!

A more useful representation is a heatmap. This will better show the relative accuracy of regions. We grouped the shots into 1ft by 1ft cells for our heatmap.

**LAKERS ACCURACY HEATMAP**

```python
# Creates colormap using 1 foot by 1 foot cells that represent shot accuracy
import numpy as np
colormap = np.zeros(shape=(47,50))
for i in range(50):
    for j in range(47):
        xmin = -250+i*10
        xmax = -240+i*10
        ymin = -50+j*10
        ymax = -40+j*10
        grid = lakers_result.loc[(lakers_result.LOC_X>=xmin) & (lakers_result.LOC_X<xmax) & (lakers_result.LOC_Y>=ymin) & (lakers_result.LOC_Y<ymax)].reset_index()
        miss = grid.loc[(grid.EVENT_TYPE == "Missed Shot")]
        make = grid.loc[(grid.EVENT_TYPE == "Made Shot")]
        # print(miss)
        # print(make)
        if(len(miss)+len(make)==0):
            pass
        else:
            colormap[j][i] = len(make)/(len(miss)+len(make))
```

```python
zvals = colormap

# import rainbow colormap
cmap = mpl.cm.get_cmap('gist_rainbow',256)

cmap._init() # create the _lut array, with rgba values
plt.clf()
plt.gcf().set_size_inches((15, 20))

# create your alpha array and fill the colormap with them.
alphas = np.linspace(0, 0.7, cmap.N+3)
cmap._lut[:,-1] = alphas

plt.imshow(court, extent=[0,500.0,0,470.0])
img = plt.imshow(zvals,cmap=cmap, origin='lower', extent=[0,500,0,470])
ax = plt.subplot()
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.5)

plt.colorbar(cax=cax)
plt.show()
```

From this data we can see that over time, the teams have become more accurate at making shots outside the 3 point line. This tracks common knowledge as there is a trend for teams in the past few years to make most of their shots from farther away. This increase in accuracy is most likely due to more training in longer shots, atheletes getting stronger, and the 3-point shot being more efficient at earning points than driving all the way into the arc or the paint. We also see that the area with some of the highest accuracy, right under the hoop, makes sense because it is quite hard to miss a layup or a dunk in the NBA.

We talked about increasing shot frequency from farther away increasing over the years, why not visualize it! We will again use a heatmap with a slightly different color bar.

**LAKERS SHOT FREQUENCY HEATMAP**

```python
# Calculate mean for shots attempted per cell excluding zeroes
tot = 0
num = 0
result = lakers_result
for i in range(50):
    for j in range(47):
        xmin = -250+i*10
        xmax = -240+i*10
        ymin = -50+j*10
        ymax = -40+j*10
        grid = result.loc[(result.LOC_X>=xmin) & (result.LOC_X<xmax) & (result.LOC_Y>=ymin) & (result.LOC_Y<ymax)].reset_index()
        if len(grid) != 0:
            tot += len(grid)
            num += 1
mean = int(tot/num)
```

```python
# Creates colormap for frequency of attempted shots
colormap = np.zeros(shape=(47,50))
totalshots = len(result)
mx = 0
for i in range(50):
    for j in range(47):
        xmin = -250+i*10
        xmax = -240+i*10
        ymin = -50+j*10
        ymax = -40+j*10
        grid = result.loc[(result.LOC_X>=xmin) & (result.LOC_X<xmax) & (result.LOC_Y>=ymin) & (result.LOC_Y<ymax)].reset_index()
        colormap[j][i] = len(grid)
        if colormap[j][i] > mean*10: 
           colormap[j][i] = mean*10
        colormap[j][i] /= len(result)
```

```python
zvals = colormap

# import rainbow colormap
cmap = mpl.cm.get_cmap('gist_rainbow',256)

cmap._init() # create the _lut array, with rgba values
plt.clf()
plt.gcf().set_size_inches((15, 20))

# create your alpha array and fill the colormap with them.
alphas = np.linspace(0.3, 0.7, cmap.N+3)
cmap._lut[:,-1] = alphas

plt.imshow(court, extent=[0,500.0,0,470.0])
img = plt.imshow(zvals,cmap=cmap, origin='lower', extent=[0,500,0,470])
ax = plt.subplot()
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.5)

plt.colorbar(cax=cax)
plt.show()
```

Aha! The heatmaps confirm common knowledge! We can see as time goes on, shots ourside of the 3 point arc become more frequent. With the Lakers we see a lot of shots in the paint and obviously some right under the hoop. We also see a lot of shots inside the 3 point arc. These inefficient 2-pointers become less and less frequent until in the Warriors' heatmap they are virtually nonexistent! The majority of the shots are either right under the hoop because of dunks and layups or 3 point shots. Its also very interesting to see the concentration of shots on the alleys being quite high. The Miami Heat were quite famous for this shot and this is shown by the high freqency of attempts from this area, even more than the Warriors!

This has some implications for modern basketball. Some people hold the opinion that this 3-point trend is boring to watch as every shot becomes essentially the same. Some have proposed the idea of moving back the 3-point arc to counteract this phenomenon. While our heatmaps don't convey any opinion about this trend, it could be an useful tool to those who want to analyze this trend in professional modern basketball.
